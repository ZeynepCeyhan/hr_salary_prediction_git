## Regression

## Setup

```python

miss=hr.isnull().sum()
miss[miss > 1]
#attrition:isten ayrilan veri sutunu
```

```python

data=hr.copy()
del data['Attrition']
```

```python
data.isnull().sum()
```

```python
missing=data[data['MonthlyIncome'].isnull()]#bos veriler
filled=data[data['MonthlyIncome'].notnull()]#dolu veriler 

```

```python
missing.shape, filled.shape
```

## SİMPLE LİNEAR

```python
filled.corr()
#dolu veriler
```

```python
abs(filled.corr()['MonthlyIncome']).sort_values(ascending=False)

#bu asamada aylikgelirin diger sutunlarla korelasyonu incelenir .%20-%90 arasi 
```

```python
from scipy import stats
est=stats.linregress(x=filled['TotalWorkingYears'],y=filled['MonthlyIncome'])#toplam kac yil calisip ne kadar almis
```

```python
est 
```

```python
est.slope #(egim, pvalue<0.05,iki sutun arasi bag var)
```

```python
est.intercept #(kesim noktasi) 
#yani kisi ise alindigi ilk ay 1187 tl aliyor=asgari maas
```

```python
def predict(x):
    return est.slope * x + est.intercept 
#iscinin alacagi maas
```

```python
predict(2) #2 yil calisan bir iscinin maasi 2 yil sonunda ne kadar olur
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(filled['MonthlyIncome'],predict(filled.TotalWorkingYears))
#%60 basari oldukca dusuk.Yukseltilmeli
```

```python
mean_squared_error(filled['MonthlyIncome'],filled['TotalWorkingYears'].apply(predict))**0.5
#2972 dolar zarar cikti
```

```python
pred=L.predict(X.drop('MonthlyIncome',axis=1))
pred
#47 sutunla fiyatlari tahmin ediyor
```
